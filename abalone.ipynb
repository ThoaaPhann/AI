```python
import pandas as pd 
url = '/content/abalone.csv'
data = pd.read_csv(url)
data.head()

import seaborn as sns 
import matplotlib.pyplot as plt 
data.hist()
plt.show()

correlation = data.corr()
print (correlation)

sns.pairplot(data)

sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x='Length', y = 'Rings',figsize = (9,9),color='blue')

from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Length']),y = data['Rings'])
print('Gia tri Intercept la:',model.intercept_)
print('Gia tri Coeffient la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X = pd.DataFrame(data['Length']),y = data['Rings']))
```

```python
data.plot(kind = 'scatter', x='Length', y = 'Rings',figsize = (9,9),color='blue')

y_pred = model.predict(X = pd.DataFrame(data['Length']))
plt.plot(data['Length'],y_pred,color = 'red')
```
