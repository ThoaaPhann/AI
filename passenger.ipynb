```python
import pandas as pd 
url = '/content/passenger.csv'
data = pd.read_csv(url)
data.head()

import seaborn as sns 
import matplotlib.pyplot as plt 
data.hist()
plt.show()

correlation = data.corr()
print (correlation)

sns.pairplot(data)

sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x='Passenger', y = 'Dollars (Y)',figsize = (9,9),color='blue')

from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Passenger']),y = data['Dollars (Y)'])
print('Gia tri Intercept la:',model.intercept_)
print('Gia tri Coeffient la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X = pd.DataFrame(data['Passenger']),y = data['Dollars (Y)']))
```

```python
data.plot(kind = 'scatter', x='Passenger', y = 'Dollars (Y)',figsize = (9,9),color='blue')

y_pred = model.predict(X = pd.DataFrame(data['Passenger']))
plt.plot(data['Passenger'],y_pred,color = 'red')
```

```python
## Bac 2 

model = linear_model.LinearRegression()
X = pd.DataFrame([data['Passenger'],data['Passenger']**2]).T 
model.fit(X,y = data['Dollars (Y)'])
print('He so Intercept la:',model.intercept_)
print('He so Coeffident la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y = data['Dollars (Y)']))

import numpy as np

range = np.arange(1.5,25,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'Passenger', y = 'Dollars (Y)', figsize=(9,9),color='blue',xlim=(0,25))
plt.plot(range, y_pred, color = 'red')
```

```python
## 2 bien

model = linear_model.LinearRegression()
model.fit(X = data.loc[:,['Passenger','Miles(X)']],y=data['Dollars (Y)'])

print('He so Coeffinet la:',model.coef_)
print('He so Intercept la:',model.intercept_)
print('Do chinh xac cua mo hinh:',model.score(X = data.loc[:,['Passenger','Miles(X)']],y=data['Dollars (Y)']))

import numpy as np

range = np.arange(1.5,25,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)

plt.figure(figsize=(9,9))
plt.scatter(x=data['Passenger'], y=data['Dollars (Y)'], color='blue', label='Passenger')
plt.scatter(x=data['Miles(X)'], y=data['Dollars (Y)'], color='green', label='Miles(X)')
plt.plot(range, y_pred, color='red')
plt.xlim((0, 25))
plt.xlabel('Passenger, Miles(X)')
plt.ylabel('Dollars (Y)')
plt.show()
```
